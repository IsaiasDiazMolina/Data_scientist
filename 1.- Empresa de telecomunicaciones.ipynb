### Datos por usuario llamadas

```python
# Calculamos el número y duracion de llamadas hechas por cada usuario por mes. 

calls_by_month = (calls.groupby(['user_id', 'month'])['duration']
                  .agg(['sum', 'count'])
                  .rename(columns={'sum': 'duration_call_sum', 'count': 'number_of_calls'})
                  .reset_index())

calls_by_month['duration_call_sum'] = np.ceil(calls_by_month['duration_call_sum'])
calls_by_month.head()
```

### Datos por usuario mensages

```python
# Calculamos el número de mensajes hechas por cada usuario por mes. 
messages_by_month = (messages.groupby(['user_id', 'month'])['id']
                    .agg('count')                     
                    .reset_index()
                    .rename(columns={'id': 'total_messages'}))
                    
messages_by_month
```

## Creando un nuevo DataFrame

```python
# Unimos los DataFrames de llamadas, mensages y volumen de datos.

total_by_month = (calls_by_month.merge(messages_by_month, on = ['user_id', 'month'], how = 'outer')
                 .merge(mb_by_month, on = ['user_id', 'month'], how = 'outer'))
```

```python
# Al nuevo DataFrame le pasamos tambien la columna plan del DataFrame users. 
total_by_month = total_by_month.merge(users[['user_id', 'plan']], on='user_id')
```

```python
# Pasamos las columnas de beneficios y pagos adicionales de cada plan.
total_by_month = total_by_month.merge(plans, how='outer', left_on='plan', right_on='plan_name')
```

## Pagos totales.

### Lamadas extra.

```python
# Creamos la columna con los minutos extra y el cobro por los minutos extra.
total_by_month['extra_minutes'] = total_by_month['duration_call_sum'] - total_by_month['minutes_included']
total_by_month.loc[total_by_month['extra_minutes'] < 0, 'extra_minutes'] = 0
total_by_month['usd_extra_minutes'] = total_by_month['extra_minutes'] * total_by_month['usd_per_minute']
```

```python
total_by_month.head()
```

Se creo una columna con los minutos extra y una mas con el pago por este extra.

### Mensages extra

```python
# Creamos la columna con los mensages extra y el cobro por los mensages extra.
total_by_month['extra_messages'] = total_by_month['total_messages'] - total_by_month['messages_included']
total_by_month.loc[total_by_month['extra_messages'] < 0, 'extra_messages'] = 0
total_by_month['usd_extra_messages'] = total_by_month['extra_messages'] * total_by_month['usd_per_message']
```

```python
total_by_month.head()
```

Se creo una columna con los mensages extra y una mas con el pago por este extra.

### Gigabytes extra

```python
# Creamos la columna con los gigabytes extra y el cobro por los gigabytes extra.
total_by_month['mb_extra'] = total_by_month['mb_used_by_month'] - total_by_month['mb_per_month_included']

total_by_month['mb_surplus'] = total_by_month['mb_extra'] / 1024

total_by_month.loc[total_by_month['mb_surplus'] < 0, 'mb_surplus'] = 0

total_by_month['mb_surplus'] = total_by_month['mb_surplus'].apply(np.ceil)

total_by_month['usd_mb_extra'] = total_by_month['mb_surplus'] * total_by_month['usd_per_gb']
```

## Pruebas de hipotesis.

Realizaremos nuestra prueba de hipotesis usando un valor alfa de .05 ya que por el tamaño de la muestra es suficiente utilizando el metodo ttest_ind ya que los datos que se usaran no dependen uno del otro y no contienen la misma cantidad de datos cada una.

Las hipotesis son:

1.-

H0: El ingreso promedio de los usuarios de las tarifas Ultimate y Surf es igual.

Ha: El ingreso promedio de los usuarios de las tarifas Ultimate y Surf difiere.

2.-

H0: El ingreso promedio de los usuarios en el área de estados Nueva York-Nueva Jersey es igual al de los usuarios de otras regiones.

Ha: El ingreso promedio de los usuarios en el área de estados Nueva York-Nueva Jersey es diferente al de los usuarios de otras regiones.

Anteriormente ya habiamos obtenido las varianzas del total de pago por cada plan, ahora sacaremos las varianzas del total de pago por ciudad para determinar que tipo de prueba debemos ocupar para determinar nuestro valor p.

```python
[np.var(x, ddof=1) for x in [total_pay_ultimate['total_pay_per_month'], total_pay_surf['total_pay_per_month']]]
```

```python
[np.var(x, ddof=1) for x in [total_pay_ny['total_pay_per_month'], total_pay_other_city['total_pay_per_month']]]
```

Tenemos varianzas diferentes en ambos casos por lo que usaremos el metodo ttest con el parametro equal_var en Falso.

```python
alpha = 0.05 

results = st.ttest_ind(total_pay_ultimate['total_pay_per_month'], total_pay_surf['total_pay_per_month'], 
                       equal_var=False) 
print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

```python
alpha = 0.05

# Prueba de igualdad de varianzas (Levene)
levene_results = st.levene(total_pay_ultimate['total_pay_per_month'], total_pay_surf['total_pay_per_month'])
print('Levene p-value:', levene_results.pvalue)

if levene_results.pvalue < alpha:
    print("Rechazamos la hipótesis nula de igualdad de varianzas en Levene")
else:
    print("No rechazamos la hipótesis nula de igualdad de varianzas en Levene")

# Ajuste del parámetro equal_var según el resultado de Levene
equal_var = True if levene_results.pvalue >= alpha else False

# Prueba t de Student con el parámetro equal_var ajustado
results = st.ttest_ind(total_pay_ultimate['total_pay_per_month'], total_pay_surf['total_pay_per_month'], equal_var=equal_var)
print('t-test p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula de igualdad de medias")
else:
    print("No rechazamos la hipótesis nula de igualdad de medias")
```
